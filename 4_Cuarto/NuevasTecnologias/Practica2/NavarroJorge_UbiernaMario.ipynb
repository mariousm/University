### Ejercicio 2

```python
#Creamos el nuevo grafo con los atributos
G = nx.Graph(nombre='red1',tipo='no-dirigida')
G.add_edge("1","2", peso=random.randint(1,10))
G.add_edge("1","3", peso=random.randint(1,10))
G.add_edge("2","3", peso=random.randint(1,10))
G.add_edge("3","6", peso=random.randint(1,10))
G.add_edge("6","4", peso=random.randint(1,10))
G.add_edge("6","7", peso=random.randint(1,10))
G.add_edge("6","5", peso=random.randint(1,10))
G.add_edge("6","8", peso=random.randint(1,10))
G.add_edge("7","10", peso=random.randint(1,10))
G.add_edge("4","5", peso=random.randint(1,10))
G.add_edge("4","8", peso=random.randint(1,10))
G.add_edge("4","9", peso=random.randint(1,10))
G.add_edge("9","5", peso=random.randint(1,10))
G.add_edge("5","8", peso=random.randint(1,10))
nx.draw(G)
```

```python
#Mostrar peso>5
for i,j in G.edges:
    if G[i][j]['peso'] > 5:
        print("Enlace: "+i+"-"+j+" peso: "+str(G[i][j]['peso']))
```

```python
#Mostrar la distribución de pesos ordenada de menor a mayor
lista = []
print("Lista sin ordenar")
for i,j in G.edges:
    print("Enlace: "+i+"-"+j+" peso: "+str(G[i][j]['peso']))
    lista.append((i,j,G[i][j]['peso']))
    
print('\n'+"Lista ordenada")
lista.sort(key=lambda x: x[2])
for i in lista:
    print("Enlace: "+i[0]+"-"+i[1]+" peso: "+str(i[2]))
```

```python
#Calcular el número de enlaces(los enlaces depende del peso) que tiene cada nodo
grados_nodos = []
for i in G.degree(weight='peso'):
    grados_nodos.append(i[1]*100)
    print("El nodo: "+i[0]+" tiene grado: "+str(i[1]))
```

```python
#Calcular los tres nodos con mayor grado
nodos_grado = []
for i in G.degree(weight='peso'):
    nodos_grado.append((i[0],i[1]))

#Ordenamos
nodos_grado.sort(key=lambda x: x[1], reverse=True)
print("Lista ordenada")
for i in nodos_grado:
    print("El nodo: "+i[0]+" tiene grado: "+str(i[1]))

print("\n3 nodos con mayor grado")
nodos_grado = nodos_grado[0:3]
for i in nodos_grado:
    print("El nodo: "+i[0]+" tiene grado: "+str(i[1]))
```

```python
#Layaout kamada_kawai
pos = nx.kamada_kawai_layout(G)
nx.draw(G, pos, with_labels=True)
nx.draw_networkx_nodes(G, pos, node_size=grados_nodos)
```

```python
#Cambiar el color de los 3 nodos de mayor grado
pos = nx.kamada_kawai_layout(G)
nx.draw(G, pos, with_labels=True)
nx.draw_networkx_nodes(G, pos, node_size=grados_nodos)
#Nos quedamos solo con el nombre del nodo [nodo[0] for nodo in nodos_grado]
#También nos quedamos con su grado [nodo[1] for nodo in nodos_grado]
nx.draw_networkx_nodes(G, pos, nodelist=[nodo[0] for nodo in nodos_grado], node_color='g',node_size=[nodo[1]*100 for nodo in nodos_grado])
```
