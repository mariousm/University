Nuestra red se caracteriza por poseer la propiedad del componente gigante ya que todos los nodos se encuentran formando un único componente, ya que no tenemos ningún artista aislado siendo nuestro trabajo analizar las colaboraciones entre artistas.

#### Propiedad de pequeño mundo

Una red de mundo pequeño es un tipo de grafo para el que la mayoría de los nodos no son vecinos entre sí, y sin embargo la mayoría de los nodos pueden ser alcanzados desde cualquier nodo origen a través de un número relativamente corto de saltos entre ellos.

Como podemos ver en la siguiente celda, lo que "nos cuesta" alcanzar un nodo es muy poco, ya que prácticamente necesitamos dos pasos para alcanzar nuestro objetivo. 

```python
# Propiedad del pequeño mundo, cuanto menor mejor, ya que significa que cuesta poco ir de un nodo a otro
suma = []
caminos = list(nx.shortest_path_length(G))
for camino in caminos:
    for valor in camino[1].values():
        suma.append(valor)
    
np.mean(suma)
```

### Centralidad de los nodos

Hemos seleccionado el pageRank para determinar la centralidad de los nodos, ya que podemos añadir la importancia de cada nodo debido a que nos hemos dado cuenta, que la red no refleja la importancia que tenga el productor de música (solo mirando las colaboraciones). Ejempo Calvin Harris.

Para comparar la centralidad de PageRank hemos usado katz_centrality ya que en ambos podemos definir una importancia, la cual va a ser la posición del productor en el top 100. De esta forma tenemos en cuenta tanto el número de colaboraciones como la posición del DJ.

Para calcular la importancia hemos indicado que el que esté en el puesto #1 tiene 1, mientras que por cada puesto que bajamos restamos 0.01, es decir, Diplo está en el puesto #42 y su importancia = ((100-42)*0.01)+0.01 = 0.59. Los que no aparecen en el top 100 su importancia es de 0.001

```python
def ordenarDiccionario (dic):
    lista = list(dic.items())
    lista.sort(reverse=True, key=lambda x:x[1])
    return lista
```

#### PageRank

En este algoritmo, si te apunta un nodo importante tu importancia va a aumentar más que si lo hace otro menos importante (siempre y cuando éste no apunte a demasiada gente, ya que si lo hace el valor que te proporciona es muy pequeño).

Este concepto trasladado a nuestro trabajo, no es lo mismo que te apunte Hardwell a que te apunte Justin Mylo.

```python
#Centralidad calculada con el pageRank, divide la importancia entre el grado
#lo cual esta es la ideal para lo que nosotros queremos analizar, si un nodo es importante y apunta a 100, no todos son importantes
centralidad = ordenarDiccionario(nx.pagerank(G, alpha=0.85, personalization=importancia))
print("CENTRALIDAD PageRank:")
print("")
print(centralidad)
```

#### Katz

El algoritmo de Katz es muy similar al pageRank, con la diferencia de que éste no divide la importancia entre el grado, es decir, que en este caso eres más importante cuanto más importante sea el nodo que te apunta.

```python
#Centralidad calculada con el pageRank, no divide la importancia entre el grado,
#no tiene en cuenta las colabroaciones con los demás, si un nodo es importante y apunta a 100, todos son importantes
centralidad = ordenarDiccionario(nx.katz_centrality(G, alpha=0.1, beta=importancia))
print("CENTRALIDAD Katz:")
print("")
print(centralidad)
```

#### CONCLUSIÓN ENTRE PAGERANK Y KATZ

Como podemos apreciar en el pageRank Robin Schulz en el pageRank está el último, ya que su única colaboración es con Guetta y éste apunta a mucho más, por lo tanto la importancia que le proporciona es menor.

Sin embargo, con Katz Robin Schulz ha subido 5 puestos, gracias a que en este caso no se divide la importancia entre el grado, es por ello que Robin adquiere más valor.

#### Centralidad Adicionales (Aunque no nos aportan mucha información)

#### Centralidad de Grado

Si solo nos fijamos en el grado de cada artista, estamos perdiendo la importancia que tiene, es decir, si haces muchas colaboraciones pero no estás dentro del top 100, realmente eres poco importante. Es por ello, que esta medida no nos proporciona toda la información necesaria para poder saber el cómo de importante es un productor.

Por ejemplo Alan Walker, en el algoritmo de Katz está en el medio (es decir es algo importante), sin embargo en la centralidad de grado, está el penúltimo.

```python
#Centralidad de grado, no tenemos en cuenta la importancia de cada productor musical, solo tenemos en cuenta las colaboraciones
centralidad = ordenarDiccionario(nx.degree_centrality(G))
print("CENTRALIDAD Grado:")
print("")
print(centralidad)
```

#### Centralidad de closeness

Nuestra red al tener la propiedad de pequeño mundo, esta medida no nos aporta mucha información, ya que todos los nodos está relativamente cerca, como hemos calculado antes tenemos un "coste" de 2.38, es decir, tardamos muy poco en ir de un nodo a otro

```python
#Centralidad de closeness
centralidad = ordenarDiccionario(nx.closeness_centrality(G))
print("CENTRALIDAD Closness:")
print("")
print(centralidad)
```

#### Centralidad de betweenness

Esta medida sí que nos aporta algo de información, ya que podemos ver los productores con los que la gente colabora más, es decir, por los que pasa más información. El incoveniente como hasta ahora, es que solo tenemos en cuenta los enlaces, no la importancia dentro del top 100.

```python
# Centralidad de betweenness
centralidad = ordenarDiccionario(nx.betweenness_centrality(G))
print("CENTRALIDAD Closness:")
print("")
print(centralidad)
```

### Comunidades

Aunque Louvain es el algoritmo más eficiente para obtener las comunidades, hemos usado  Girvan‐Newman ya que éste se encuentra directamente en networkx.

Como podemos apreciar en la siguiente celda, el algoritmo de Girvan-Newman nos devuelve 8 comunidades, a excepción de las comunidades en la que hay artistas solitarios como Diplo, Alan Walker y Robin Schulz, en las demás se caracteriza porque tienen estilos similar entre los artistas que pertenecen a la comunidad.

Aún así esta división no es del todo buena, es por ello que hemos decidido aplicar el agoritmo de Louvain en Gephi y así obtener una mejor división de comunidades.

#### Girvan-Newman

```python
communities_generator = nx.algorithms.community.girvan_newman(G)
partitions = []
modularity = []
for part in communities_generator:
    partitions.append(part)
    modularity.append(nx.community.modularity(G,part))
best_p=np.argmax(np.array(modularity))
mejor_particion = partitions[best_p]
print('Número de comunidades: %i' % len(mejor_particion))
print("")
print('Modularidad: %0.6f' % nx.community.modularity(G,mejor_particion))
print("")

for i in range(len(mejor_particion)):
    print('Partición ' + str(i+1) + "  ->  " + str(mejor_particion[i]))
    
# COLORES PARA EL GRAFO
colores = ["blue", "orange", "green", "red", "purple", "brown", "pink", "gray", "olive", "cyan"]
dic_colores = {}
nodos_colores = []

for particion in range(len(mejor_particion)):
    for nodo in G.nodes():
        if nodo in mejor_particion[particion]:
            dic_colores[nodo] = colores[particion]
            
#Para que esté en el orden según G.nodes()
for nodo in G.nodes():
    nodos_colores.append(dic_colores[nodo])
```

```python
#Layaout kamada_kawai
plt.figure(figsize=(21,21))
pos = nx.kamada_kawai_layout(G)
nx.draw(G, pos, with_labels=True)
nx.draw_networkx_nodes(G, pos, node_size=grados_nodos, node_color = nodos_colores)
```

### Distribución de grado

Respecto a la distribución de grado

```python
def gradosNodos(G):
    grados_nodos = []
    for nodo in G.nodes():
        grados_nodos.append(G.degree(nodo))
    return grados_nodos
```

```python
plt.figure(figsize=(10,5))
plt.hist(gradosNodos(G), bins = 10)
```

```python
# Logaritmica
plt.figure(figsize=(10,5))
clases = np.array([2**x for x in range(0,4)])
rango = clases[1:]-clases[:-1]
histograma, bin_enlaces =np.histogram(np.array(gradosNodos(G)),bins=clases)
histograma=histograma/rango
plt.bar(bin_enlaces[:-1],histograma,width=rango, alpha=0.5,align='edge')
plt.yscale('log');
plt.xscale('log');
```

```python
# Calculada la regresión
pendiente, intercept, correlacion, pvalor, error = stats.linregress(np.log(bin_enlaces[0:-1]),np.log(histograma))
print("Pendiente: " + str(pendiente))
print("El P-Valor: " + str(pvalor))
print("Coeficiente de correlación: "+ str(correlacion))
```
